```python
from ultralytics import RTDETR
model = RTDETR('rtdetr-l.pt')

# Define paths
data_path = 'dataset.yaml'
save_model_path = './models/yolov8-v2.pt'

# Train the model with augmentations
model_results = model.train(
    data=data_path,
    epochs=15,
    imgsz=640,
    augment=True,  # Enable augmentations
)


# Save the trained model
model.save(save_model_path)
```

```python
from ultralytics import RTDETR

model_path = "./runs/detect/train95/weights/best.pt"
model = RTDETR(model_path)

def predict(image_paths):
    # Load the pretrained or fine-tuned YOLOv8 model
      # Path to your fine-tuned model

    # Predict on the images using the GPU (CUDA)
    results = model(source=image_paths, device='cuda', save=True, save_txt=True)

    return results
```

```python
import os

# Directory containing the images
image_dir = "./val"

# List all image files in the directory
image_paths = [os.path.join(image_dir, img) for img in os.listdir(image_dir) if img.endswith(('.jpg', '.jpeg', '.png'))][:10]
```

```python
results = predict(image_paths)
```

```python
results[0].boxes.xyxy
```

```python
from PIL import Image, ImageDraw, ImageFont

idx=1

# Load the image
image_path = image_paths[idx]
image = Image.open(image_path)

# Process the results
for result in results[idx]:
    boxes = result.boxes.xyxy  # Get bounding boxes (x1, y1, x2, y2)


    # Draw bounding boxes for the desired class
    draw = ImageDraw.Draw(image)
    font = ImageFont.load_default()  # Load default font
    for box in boxes:
        x1, y1, x2, y2 = box
        draw.rectangle([x1, y1, x2, y2], outline="green", width=2)
        draw.text((x1, y1 - 10), "object", fill="green", font=font)

# Display the image
image
```
